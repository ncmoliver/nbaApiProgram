```python
player_stats_url = f"https://api.sportsdata.io/v3/nba/stats/json/PlayerSeasonStats/2024?key={nba_api_key}"
stat_response = requests.get(player_stats_url)
stat_response = stat_response.json()
print(json.dumps(stat_response, indent=4))
```

```python
player_stats_df = pd.json_normalize(stat_response)
player_stats_df.columns
```

```python
columns_selected = ["Name", "Team", "Position", "Games", "Minutes", "TwoPointersMade", "TwoPointersAttempted", "ThreePointersMade", "ThreePointersAttempted","FreeThrowsMade", "FreeThrowsAttempted", "OffensiveRebounds", "DefensiveRebounds", "Rebounds", "Assists", "Steals", "BlockedShots", "Turnovers", "PersonalFouls", "Points"]
new_player_stat_df = player_stats_df[columns_selected].copy()
new_player_stat_df.head()
```

```python
def missed_shots(attempted, made):
    return attempted - made

def calculate_stats(df):
    df.loc[:, "Min per Game"] = round(df["Minutes"] / df["Games"], 2)
    
    # Calculate missed shots for two-pointers and three-pointers
    df.loc[:, "MissedTwoPointers"] = df.apply(lambda row: missed_shots(row["TwoPointersAttempted"], row["TwoPointersMade"]), axis=1)
    df.loc[:, "MissedThreePointers"] = df.apply(lambda row: missed_shots(row["ThreePointersAttempted"], row["ThreePointersMade"]), axis=1)
    df.loc[:, "MissedFreeThrows"] = df.apply(lambda row: missed_shots(row["FreeThrowsAttempted"], row["FreeThrowsMade"]), axis=1)
    
    # Calculate two-point shot production (made shots over attempts)
    df.loc[:, "TwoShotProduction"] = df["TwoPointersMade"] - df["MissedTwoPointers"].replace(0, 1)  # Avoid division by zero
    
    # Calculate three-point shot production (made shots over attempts)
    df.loc[:, "ThreeShotProduction"] = df["ThreePointersMade"] - df["MissedThreePointers"].replace(0, 1)  # Avoid division by zero
    
    # Calculate free-throw shot production (made shots over attempts)
    df.loc[:, "FreeThrowProduction"] = df["FreeThrowsMade"] - df["MissedFreeThrows"].replace(0, 1)  # Avoid division by zero
    
    return df
```

```python
calculated_df = calculate_stats(new_player_stat_df)
calculated_df.head()
```

```python
calculated_df.columns
```

```python
# Calculate Rebounds
def calculate_aux(df):
    df.loc[:,"OR_Production"] = df["OffensiveRebounds"] * 1
    df.loc[:,"DR_Production"] = df["DefensiveRebounds"] * 0.50
    df.loc[:, "Stl_Production"] = df["Steals"] * 0.75
    df.loc[:, "Ass_Production"] = df["Assists"] * 0.50
    df.loc[:, "Blk_Production"] = df["BlockedShots"] * 0.50
    return df
    
```

```python
updated_aux_df = calculate_aux(calculated_df)
updated_aux_df.head()
```

```python
select_columns = ["Name", "Team", "Position", "Min per Game", "TwoShotProduction", "ThreeShotProduction", "FreeThrowProduction", "OR_Production", "DR_Production",	"Stl_Production",	"Ass_Production",	"Blk_Production" ]
processed_df = updated_aux_df[select_columns]
processed_df.groupby(by="Team").head()
```
